# **EXPLORATORY DATA ANALYSIS**

----- NO. OF CUSTOMERS BY GENDER -----

```python
a = sns.countplot(x = 'Gender', data = df)
for i in a.containers:
    a.bar_label(i)
```

----- SPENDITURE BY GENDER -----

```python
b = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Gender', y = 'Amount', data = b)
```

CONCLUSION 1 : Most of the buyers are Female and even the purchasing power of female is greater than male

----- NO. OF CUSTOMERS BY AGE GROUP -----

```python
z = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
for i in z.containers:
    z.bar_label(i)
```

----- SPENDITURE BY AGE GROUP -----

```python
x = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Age Group', y = 'Amount', data = x)
```

CONCLUSION 2 : From above graphs we can see that most of the buyers are of age group between 26-35 yrs female

----- ORDERS AND SALES FROM STATES -----

```python
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc = {'figure.figsize' : (17,5)})
sns.barplot(data = sales_state, x = 'State', y = 'Orders')
```

```python
amount_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(data = amount_state, x = 'State', y = 'Amount')
```

CONCLUSION 3 : From abouve graphs we can see that most of the orders and total sales/amount are from Uttar Pradesh, Maharastra and Karnataka respictively, whereas 'Kerala' having higher Orders but low Sales and on the other hand 'Bihar' is not in the list of Orders but they generated decent Amount

----- ORDERS AND SALES 
BY MARITAL STATUS -----

```python
ax = sns.countplot(data = df, x = 'Marital_Status')
sns.set(rc={'figure.figsize':(3,5)})
for i in ax.containers:
    ax.bar_label(i)
```

```python
amount_marital = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(data = amount_marital, x = 'Marital_Status', y = 'Amount', hue = 'Gender')
```

CONCLUSION 4 : From the above graph we can see that most of the buyers are married (women) and they have high purchasing power

----- ORDERS AND SALES BY OCCUPATION -----

```python
sns.set(rc={'figure.figsize':(20,5)})
a = sns.countplot(data = df, x = 'Occupation')

for i in a.containers:
    a.bar_label(i)
```

```python
sales = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(data = sales, x = 'Occupation', y = 'Amount')
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector and also they are spending more

----- ORDERS AND SALES BY PRODUCT CATEGORY -----

```python
sns.set(rc={'figure.figsize':(25,10)})
a = sns.countplot(data=df, x='Product_Category')
for i in a.containers:
    a.bar_label(i)
```

```python
sns.set(rc={'figure.figsize':(25,10)})
b = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(data=b, x='Product_Category', y='Amount')
```

From above graphs we can see that most of the sold products are from Clothing & Apparel, Food, and Electronics Category but Amount generated is mostly from Food then Clothing & Apparel and Electronics Category

----- ORDERS AND SALES BY PRODUCTS -----

```python
sales_products = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_products, x = 'Product_ID', y = 'Orders')
```

FINAL CONCLUSION : Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics Category
